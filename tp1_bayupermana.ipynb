```python
# Tugas Personal 1 - Data Science
import pandas as pd

# Mengambil data karyawan melalui file .csv
data = pd.read_csv("data_karyawan.csv", index_col=0)
df = pd.DataFrame(data)

# Menentukan jabatan berdasarkan gaji
jabatan = []
for gaji in df['Gaji']:
    if 8000000 <= gaji <= 9000000:
        jabatan.append("Officer")
    elif 10000000 <= gaji <= 11000000:
        jabatan.append("Supervisor")
    elif 12000000 <= gaji <= 14000000:
        jabatan.append("Asisten Manajer")
    else:
        jabatan.append("Manager")

df['Jabatan'] = jabatan

# Mengurutkan gaji terbesar ke gaji terkecil
df = df.sort_values(by='Gaji', ascending=False)

# Menampilkan seluruh data table
df
```
